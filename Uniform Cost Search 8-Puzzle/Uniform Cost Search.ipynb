```python
import numpy as np
from queue import PriorityQueue
from copy import deepcopy
import queue
import heapq
Initialpuzzle = [2,4,6,8,1,3,5,7,0]
goal = [1,2,3,4,5,6,7,8,0]
```

Classes and ultility functions

```python
class Node:
    def __init__(self, P_state,_parent,_cost = 0):
        self.state = P_state
        self.parent = _parent
        self.cost = _cost;
    def Getcost(self):
        return self.cost;
    def __lt__(self,other):
        return self.cost < other.cost

def CanIMoveLeft(state):
    blank = state.index(0);
    if blank in [0,3,6]:
        return False;
    else:
        return True;
def CanIMoveRight(state):
    blank = state.index(0);
    if blank in [2,5,8]:
        return False;
    else:
        return True;
def CanIMoveUp(state):
    blank = state.index(0);
    if blank in [0,1,2]:
        return False;
    else:
        return True;
def CanIMoveDown(state):
    blank = state.index(0);
    if blank in [6,7,8]:
        return False;
    else:
        return True;


def MoveUp(state):
    new_state = deepcopy(state)
    blank = new_state.index(0);
    if blank not in [0,1,2]:
        ind = blank-3
        tmp = new_state[ind]
        new_state[ind] = new_state[blank]
        new_state[blank] = tmp
        return new_state
    else:
        return state
def MoveDown(state):
    new_state = deepcopy(state)
    blank = new_state.index(0);
    if blank not in [6,7,8]:
        ind = blank+3
        tmp = new_state[ind]
        new_state[ind] = new_state[blank]
        new_state[blank] = tmp
        return new_state
    else:
        return state
def MoveLeft(state):
    new_state = deepcopy(state)
    blank = new_state.index(0);
    if blank not in [0,3,6]:
        ind = blank-1
        tmp = new_state[ind]
        new_state[ind] = new_state[blank]
        new_state[blank] = tmp
        return new_state
    else:
        return state
def MoveRight(state):
    new_state = deepcopy(state)
    blank = new_state.index(0);
    if blank not in [2,5,8]:
        ind = blank+1
        tmp = new_state[ind]
        new_state[ind] = new_state[blank]
        new_state[blank] = tmp
        return new_state
    else:
        return state


def Move(state):
    NewStates = []
    NewStates.append(MoveUp(state))
    NewStates.append(MoveDown(state))
    NewStates.append(MoveLeft(state))
    NewStates.append(MoveRight(state))
    return NewStates
def TracePath(x):
    path = []
    while(x.parent!=[]):
        path.append(x.state)
        x = x.parent
    return path
def printPuzzle(state):
    print(state[0:3])
    print(state[3:6])
    print(state[6:9])


def convetToMatrix(state):
    statecopy = deepcopy(state)
    ret = []
    ret.append(statecopy[0:3])
    ret.append(statecopy[3:6])
    ret.append(statecopy[6:9])
    return ret
 
```

Uniform Cost Search Algo


```python


def UCS(puzzle, goal):
    Q = PriorityQueue()
    isvisited = []
    start = Node(puzzle,[],0); #starting with cost 0
    Q.put((start.cost,start));
    isvisited.append(start.state);
    MoveList = []
    i = 0
    while True:
        i+=1
        if i%100 == 0:
            print("Still Searching...")
        x = Q.get();
        if CanIMoveUp(x[1].state) == True:
            n1 = Node(MoveUp(x[1].state), x[1], int(x[1].cost)+1)
            if n1.state not in isvisited:
                isvisited.append(n1.state)
                Q.put((n1.cost, n1))

        if CanIMoveDown(x[1].state) == True:
            n2 = Node(MoveDown(x[1].state), x[1], int(x[1].cost)+1)
            if n2.state not in isvisited:
                isvisited.append(n2.state)
                Q.put((n2.cost, n2))

        if CanIMoveLeft(x[1].state) == True:
            n3 = Node(MoveLeft(x[1].state), x[1], int(x[1].cost)+1)
            if n3.state not in isvisited:
                isvisited.append(n3.state)
                Q.put((n3.cost, n3))

        if CanIMoveRight(x[1].state) == True:
            n4 = Node(MoveRight(x[1].state), x[1], int(x[1].cost)+1)
            if n4.state not in isvisited:
                isvisited.append(n4.state)
                Q.put((n4.cost, n4))
        if(x[1].state == goal):
            print("Got it...")
            return TracePath(x);
```
