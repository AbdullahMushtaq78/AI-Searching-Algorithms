```python
import numpy as np
from queue import PriorityQueue
from copy import deepcopy
import queue
import heapq
Initialpuzzle = [2,4,6,8,1,3,5,7,0]
goal = [1,2,3,4,5,6,7,8,0]
```

Classes and ultility functions

```python
class Node:
    def __init__(self, P_state,_parent,_cost = 0):
        self.state = P_state
        self.parent = _parent
        self.cost = _cost;
    def Getcost(self):
        return self.cost;
    def __lt__(self,other):
        return self.cost < other.cost

def CanIMoveLeft(state):
    blank = state.index(0);
    if blank in [0,3,6]:
        return False;
    else:
        return True;
def CanIMoveRight(state):
    blank = state.index(0);
    if blank in [2,5,8]:
        return False;
    else:
        return True;
def CanIMoveUp(state):
    blank = state.index(0);
    if blank in [0,1,2]:
        return False;
    else:
        return True;
def CanIMoveDown(state):
    blank = state.index(0);
    if blank in [6,7,8]:
        return False;
    else:
        return True;


def MoveUp(state):
    new_state = deepcopy(state)
    blank = new_state.index(0);
    if blank not in [0,1,2]:
        ind = blank-3
        tmp = new_state[ind]
        new_state[ind] = new_state[blank]
        new_state[blank] = tmp
        return new_state
    else:
        return state
def MoveDown(state):
    new_state = deepcopy(state)
    blank = new_state.index(0);
    if blank not in [6,7,8]:
        ind = blank+3
        tmp = new_state[ind]
        new_state[ind] = new_state[blank]
        new_state[blank] = tmp
        return new_state
    else:
        return state
def MoveLeft(state):
    new_state = deepcopy(state)
    blank = new_state.index(0);
    if blank not in [0,3,6]:
        ind = blank-1
        tmp = new_state[ind]
        new_state[ind] = new_state[blank]
        new_state[blank] = tmp
        return new_state
    else:
        return state
def MoveRight(state):
    new_state = deepcopy(state)
    blank = new_state.index(0);
    if blank not in [2,5,8]:
        ind = blank+1
        tmp = new_state[ind]
        new_state[ind] = new_state[blank]
        new_state[blank] = tmp
        return new_state
    else:
        return state


def Move(state):
    NewStates = []
    NewStates.append(MoveUp(state))
    NewStates.append(MoveDown(state))
    NewStates.append(MoveLeft(state))
    NewStates.append(MoveRight(state))
    return NewStates
def TracePath(x):
    path = []
    while(x.parent!=[]):
        path.append(x.state)
        x = x.parent
    return path
def printPuzzle(state):
    print(state[0:3])
    print(state[3:6])
    print(state[6:9])


def convetToMatrix(state):
    statecopy = deepcopy(state)
    ret = []
    ret.append(statecopy[0:3])
    ret.append(statecopy[3:6])
    ret.append(statecopy[6:9])
    return ret

def MisplacedTiles(state):
    count = 0
    for i in range(9):
        if state[i] != goal[i]:
            count+=1
    return count

def Manhattan(state):
    statecopy = deepcopy(state)
    mat = convetToMatrix(statecopy)
    ideal = {1:[0,0], 2:[0,1], 3:[0,2],
             4:[1,0], 5:[1,1], 6:[1,2],
             7:[2,0], 8:[2,1], 0:[2,2]
            }
    h = 0
    for i in range(3):
        for j in range(3):
            m = ideal.get(mat[i][j])
            if m != [i,j]:
                h += abs(m[0]-i) + abs(m[1]-j)
    return h

 
```

Greedy Best First Search Algo


```python
def GBFS(puzzle, goal):
    Q = PriorityQueue()
    h = MisplacedTiles(puzzle)

    start = Node(puzzle,[],0)
    Q.put((h,start));
    isvisited = []
    isvisited.append(start.state)
    while True :
        x = Q.get()
        h = MisplacedTiles(x[1].state)
        if(h <= 0):
            return TracePath(x[1]);
        if CanIMoveUp(x[1].state) == True:
            n1 = Node(MoveUp(x[1].state), x[1], int(x[1].cost)+1)
            if n1.state not in isvisited:
                isvisited.append(n1.state)
                Q.put((MisplacedTiles(n1.state), n1))

        if CanIMoveDown(x[1].state) == True:
            n2 = Node(MoveDown(x[1].state), x[1], int(x[1].cost)+1)
            if n2.state not in isvisited:
                isvisited.append(n2.state)
                Q.put((MisplacedTiles(n2.state), n2))

        if CanIMoveLeft(x[1].state) == True:
            n3 = Node(MoveLeft(x[1].state), x[1], int(x[1].cost)+1)
            if n3.state not in isvisited:
                isvisited.append(n3.state)
                Q.put((MisplacedTiles(n3.state), n3))

        if CanIMoveRight(x[1].state) == True:
            n4 = Node(MoveRight(x[1].state), x[1], int(x[1].cost)+1)
            if n4.state not in isvisited:
                isvisited.append(n4.state)
                Q.put((MisplacedTiles(n4.state), n4))
```

```python
def GBFSMAIN():
    path = GBFS(Initialpuzzle, goal)
    path.reverse()
    print("No of Moves to reach Goal: ", len(path))
    print(">>>>>>>>>>>>>>>>>>>>Path Taken<<<<<<<<<<<<<<<<<<<<<\n\n\n")

    for i in range(len(path)):
        printPuzzle(path[i])
        if i-1 < len(path):
            print('    |\n    |\n    v')


if __name__ == "__main__":
    
    Initialpuzzle  = []
    for i in range(9):
        val = int(input("Enter values: "))
        if val not in Initialpuzzle and val <9 and val >=0:
            Initialpuzzle.append(val)
        else:
            print("This number is not allowed in this game.")
    print("Given Puzzle is: \n",Initialpuzzle)
GBFSMAIN()
```
