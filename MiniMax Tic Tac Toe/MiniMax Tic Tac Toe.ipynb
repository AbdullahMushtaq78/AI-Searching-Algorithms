```python
#Min-Max Algorithm
from math import inf
from random import choice
import platform
import numpy as np
from os import system
```

```python
Human = -1
Computer = +1
draw = 0
board = np.zeros((3,3),dtype=np.int32)
```

```python



def evaluate(curr_state):
    #write your code here to assign score if computer or human wins and 0 in case of draw; return score 
    if wins(curr_state,Human):
        return -1
    elif wins(curr_state,Computer):
        return +1
    else:
        return 0



def wins(state, player):
    win_prob = []
    #return True or False by checking all possibilites of win for human or computer
    for i in range(3):
        temp = []
        for j in range(3):
            temp.append(state[i][j])
        win_prob.append(temp)
    for i in range(3):
        temp = []
        for j in range(3):
            temp.append(state[j][i])
        win_prob.append(temp)
    temp = []
    for i in range(3):
        temp.append(state[i][i])
    win_prob.append(temp)

    temp = []
    temp.append(state[2][0])            
    temp.append(state[1][1])            
    temp.append(state[0][2])    
    win_prob.append(temp)
    if [player,player,player] in win_prob:
        return True
    else:
        return False        
def game_over(curr_state):
    # return True or False if human or computer wins
    Hw = wins(curr_state, Human)
    Cw = wins(curr_state, Computer)
    return Hw or Cw

def empty_cells(curr_state):
    # create a list of all empty cells[x,y] and return emptyCells list
    blank = list()
    for i,row in enumerate(curr_state):
        for j,tile in enumerate(row):
            if not tile:
                blank.append([i,j]);
    return blank

def valid_move(x, y):
    # check valid move if x,y is in emptyList return true else return false
    FreeTiles = empty_cells(board)
    if [x,y] in FreeTiles:
        return True
    return False
    

def set_move(x, y, player):
    # check selected cell[x,y]; if it is valid move then place player move on the board
    if valid_move(x,y):
        board[x][y] = player
        return True
    return False
    
def minimax(curr_state, Tree_depth, player):
    #Implement min-max algorithm here with following inputs:
    # Input:current-state of board, position of state where depth=[0-9], also player:human or computer
    #Output: list =[best row,best column,best score]
    if player == Computer:
        Max = [-1,-1,-inf]
    else:
        Max = [-1,-1,+inf]
    
    if Tree_depth == 0 or game_over(curr_state):
        ret_scr = evaluate(curr_state)
        return [-1,-1,ret_scr]
    
    for tile in empty_cells(curr_state):
        x,y = tile[0],tile[1]
        curr_state[x][y] = player
        ret_scr = minimax(curr_state, Tree_depth-1, -player)
        curr_state[x][y] = 0
        ret_scr[0], ret_scr[1] = x,y

        #Setting up the main algo here... Maximizing the error values for User/human and Minimizing the error Values for the AI
        if player == Computer:
            if ret_scr[2] > Max[2]:
                Max = ret_scr
        else:
            if ret_scr[2] < Max[2]:
                Max = ret_scr
    return Max

def clean():
    os_name = platform.system().lower()
    if 'windows' in os_name:
        system('cls')
    else:
        system('clear')


def printBoard(curr_state, comp_choice, human_choice):
    #print board e.g.
    # ---------------
    # |   ||   ||   |
    # ---------------
    # |   ||   ||   |
    # ---------------
    # |   ||   ||   |
    # ---------------

    chars = {
        -1: human_choice,
        +1: comp_choice,
        0: ' '
    }
    str_line = '---------------'

    print('\n' + str_line)
    for row in curr_state:
        for cell in row:
            symbol = chars[cell]
            print(f'| {symbol} |', end='')
        print('\n' + str_line)


def computer_turn(comp_choice, human_choice):
    # comp_choice: X or 0, human_choice: X or 0 for depth <9
    Tree_depth = len( empty_cells(board))
    if Tree_depth ==0 or game_over(board):
        return None
    clean()
    print("Computer turn.")
    printBoard(board, comp_choice, human_choice)
    move = minimax(board, Tree_depth, Computer)
    set_move(move[0], move[1], Computer)

def human_turn(comp_choice, human_choice):
    # comp_choice: X or 0, human_choice: X or 0 for depth <9
    ActionsMap = {
        0: [0, 0], 1: [0, 1], 2: [0, 2],
        3: [1, 0], 4: [1, 1], 5: [1, 2],
        6: [2, 0], 7: [2, 1], 8: [2, 2],
        }

    Tree_depth = len(empty_cells(board))
    if Tree_depth ==0 or game_over(board):
        return None
    clean()
    print("Your Turn: ")
    printBoard(board,comp_choice, human_choice)
    while True:
        action = input("Enter the Move (Box-0 --> Box-8):\t")
        if action.isdigit() and int(action) in range(9):
            action = int(action)
            coord = ActionsMap[action]
            if set_move(coord[0],coord[1], Human):
                break
            else:
                print("Wrong Choice...")
        else:
            print("Wrong Choice...")
    
    
```

```python

def ShowWinner(agent_Char, human_Char):
    if wins(board, Human):
        print("You Win!")
        printBoard(board, agent_Char, human_Char)
    elif wins(board, Computer):
        print("Machine Uprisal!...")
        printBoard(board, agent_Char, human_Char)
    else:
        print("Draw...")
        printBoard(board, agent_Char, human_Char)    
    
def main():
    human_Char = 'X'
    agent_Char = 'O'
    while len(empty_cells(board)) > 0 and not game_over(board):
        human_turn(agent_Char, human_Char)
        computer_turn(agent_Char, human_Char)
    ShowWinner(agent_Char, human_Char)
    #Start Working to call other functions here 
        


```
